```python
# Definimos una clase base llamada Animal
class Animal:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def hacer_sonido(self):
        return "Sonido gen√©rico"

    def descripcion(self):
        return f"{self.nombre} tiene {self.edad} a√±os."
```

```python
# Creamos clases hijas que heredan de Animal
class Perro(Animal):
    def hacer_sonido(self):
        return "Guau üê∂"


class Gato(Animal):
    def hacer_sonido(self):
        return "Miau üê±"
```

```python
perro = Perro("Firulais", 5)
gato = Gato("Michi", 3)

# Usamos sus m√©todos
print(perro.descripcion())   # Firulais tiene 5 a√±os.
print(perro.hacer_sonido())  # Guau üê∂

print(gato.descripcion())    # Michi tiene 3 a√±os.
print(gato.hacer_sonido())   # Miau üê±
```
