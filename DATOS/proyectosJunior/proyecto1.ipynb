```python
import requests
import pandas as pd

def binance_klines(symbol="BTCUSDT", interval="1d", limit=1):
    url = "https://api.binance.com/api/v3/klines"
    params = {"symbol": symbol, "interval": interval, "limit": limit}
    resp = requests.get(url, params=params, verify=False)  # ðŸ‘ˆ clave
    data = resp.json()
    print(resp)
    return pd.DataFrame(data)

df = binance_klines()
print(df.head())

```
